### Q13. Create New `income_tax` column

An **income tax** is imposed on the annual income of all registerants. The tax is calculated as follows:

- if the annual income is $\$65,000$ or less, the registrant does not need to pay tax
- if the  annual income is more than $\$65,000$ the registrant needs to pay a tax equal to $7.5\%$ of his/her annual income

Mathematically, for an registrant with an income of $\$annual\_income$, the amount of tax the registrant needs to pay, or the $\$tax$ is:

$$
tax = \begin{cases}
0 & \text{ if } annual\_income \leq 65000\\
0.075 \times annual\_income & \text{ if } annual\_income > 65000\\\end{cases}
$$

You've been tasked to create a new column called `income_tax` based on the above conditions. 

```python
# Debug calculate_income_tax(annual_inc) so that the correct tax is calculated and returned
def calculate_income_tax(annual_inc):
    if annual_inc <= 65000:
        return(0)
    else:
        return(annual_inc * 0.075)
```

```python
loans_relevant_df.dtypes
```

**Expected Output**:

<img src="https://i.imgur.com/u54cP8a.png">

```python
loans_relevant_df['income_tax'] = loans_relevant_df['annual_inc'].apply(calculate_income_tax)
loans_relevant_df.head()
```

### Q20. Outlier Analysis

Your project manager wants to investigate if the large proportions of the outliers based on `annual_inc` are `Charged Off`. First get the 25th and 75th percentile, then filter based on the interquartile range. Extract these outliers and filter them for `Charged Off` and get the proportion of these outliers with and without `Charge Off`.

**Expected Output**:

    0.08832807570977919

```python
# get percentiles 
quartile_1 = loans_relevant_df['annual_inc'].quantile(q=0.25)
quartile_3 = loans_relevant_df['annual_inc'].quantile(q=0.75)
print(quartile_1, quartile_3)

# iqr 
IQR = quartile_3 - quartile_1
print(IQR)

# calculate bounds
upper_bound = quartile_3 + 1.5 * IQR
print(upper_bound)

lower_bound = quartile_1 - 1.5 * IQR
print(lower_bound)

# filter for outliers
high_cond = loans_relevant_df['annual_inc'] > upper_bound
low_cond = loans_relevant_df['annual_inc'] < lower_bound

# need to be |, cannot be both at the same time
outliers_df = loans_relevant_df[high_cond | low_cond]
print(len(outliers_df)) # 951 
```

```python
# filter them for Charged Off
outliers_charged_off = outliers_df[outliers_df['loan_status'].isin(['Charged Off'])]
outliers_charged_off

# filter them for without Charged Off
outliers_without = outliers_df[outliers_df['loan_status'].isin(['Charged Off']) == False]
outliers_without
```

```python
# get the proportion of these outliers with and without Charge Off
total_outliers = len(outliers_charged_off) + len(outliers_without)
total_outliers    #951
proportion = len(outliers_charged_off) / total_outliers
proportion
```
